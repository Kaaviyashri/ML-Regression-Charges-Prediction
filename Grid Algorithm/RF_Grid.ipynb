```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset, dtype = int, drop_first = True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent = dataset[['charges']]
dependent
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'criterion':['squared_error', 'absolute_error'],
              'max_features':[None, 'sqrt', 'log2'],
              'n_estimators':[10,100]}
grid = GridSearchCV(RandomForestRegressor(), param_grid, refit = True, verbose = 3,  n_jobs = -1)
grid.fit(independent, dependent)
```

```python
re = grid.cv_results_
print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
table = pd.DataFrame.from_dict(re)
```

```python
table
```

```python
age_input = float(input("Age: "))
bmi_input = float(input("BMI: "))
children_input = int(input("Children: "))
sex_male_input = int(input("Sex Male - 1 for yes or 0 for no: "))
smoker_yes_input = int(input("Smoke Yes - 1 for yes or 0 for no: "))
```

```python
result = grid.predict([[age_input, bmi_input, children_input, sex_male_input, smoker_yes_input]])
print("result = {}".format(result))
```
